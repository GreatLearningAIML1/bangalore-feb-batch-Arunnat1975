### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

#### Collect Fashion mnist data from tf.keras.datasets 

```python
#import MNIST dataset
from keras.datasets import mnist
```

```python
#Load the dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
#Finding the unique values of the target variable
np.unique(y_train)
```

```python
# X_train has values up to 255. this needs to be normalised
np.unique(X_train)
```

#### Change train and test labels into one-hot vectors & normalize data

```python
#Normalising X_train and X_test
X_train = X_train/255
X_test = X_test/255
```

```python
#checking if normalisation has worked
np.unique(X_train)
```

```python
# one hot encode of test and train labels
# We will have 10 classes for the 10 digits
from keras.utils import np_utils
number_of_classes = 10
y_train = np_utils.to_categorical(y_train, number_of_classes)
y_test = np_utils.to_categorical(y_test, number_of_classes)
```

```python
#Checking the one hot encoded matrix to see if it is done
y_train[0:1]
```

#### Build the Graph

#### Initialize model, reshape

```python
# 28*28 = 784
X_train = X_train.reshape(60000,784) 
X_test = X_test.reshape(10000,784)
```

```python
X_train.shape, X_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.activations import relu,softmax
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
#Initialize Sequential Graph (model)
model = Sequential()
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```python
#Add first Dense Layer which provides 200 Outputs after applying relu
model.add(Dense(200, activation='relu',input_dim=784)) 

#Add second Dense Layer which provides 100 Outputs after applying relu
model.add(Dense(100, activation='relu'))

# Add a drop out layer p=0.25
model.add(Dropout(0.25))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```python
# Output layer with 10 neuraons for the 10 classes and softmax activation
model.add(Dense(10, activation='softmax'))
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Fit the model
model.fit(X_train,y_train,epochs=7,verbose=2)
```

```python
eval1_list=[]
eval1_list = model.evaluate(X_test, y_test)
print('Model Loss is',eval1_list[0])
print('Model Accuracy is',eval1_list[1])
```
